```python
from scipy.io import wavfile
import numpy as np
import pysptk
import matplotlib.pyplot as plt
```

```python
sound = wavfile.read('data_raw/wav/'+'arctic_a0001'+'.wav')[1]
sound = np.array(sound, dtype=float)
sounds = []
sounds.append(sound)
delete_numbers = []
# lepes merete 5ms 16kHz -> 80
for i in range(0,80):
    delete_numbers.append(i)
for i in range(0,4):
    sounds.append(np.delete(sounds[-1], delete_numbers, None))
```

```python
f0s = []
for s in sounds:
    # abalak merete 25ms 16kHz -> 400
    f0s.append(pysptk.sptk.swipe(x=sound, fs=16000.0, hopsize=400, otype='f0'))
```

```python
f0s = np.asarray(f0s).flatten('F')
```

```python
plt.plot(f0s)
plt.show()
```

```python
print(len(sound))
print(len(f0s))
print(len(sound)/len(f0s))
```
